# DataFrame

```python
import pandas as pd
# import numpy as np
from numpy.random import randn
```

### 1. Criação de um dataframe

```python
dataframe = pd.DataFrame(randn(5, 4), index=["A", "B", "C", "D", "E"], columns="W X Y Z".split())
```

```python
dataframe
```

```python
type(dataframe)
```

### 2. Seleção de uma series do dataframe

```python
dataframe["W"]
```

```python
type(dataframe["W"])
```

### 3. Seleção de uma series no formato dataframe 

```python
dataframe[["W"]]
```

```python
type(dataframe[["W"]])
```

### 4. Seleção de uma ou mais series

```python
dataframe[["W", "X"]]
```

```python
type(dataframe[["W", "X"]])
```

### 5. Manipulações no DataFrame

##### 1. Criação de uma coluna que é a cópia de uma já existente

```python
dataframe["new"] = dataframe["W"]
```

```python
dataframe
```

##### 2. Criação de uma coluna que é a soma de outras duas

```python
dataframe["sum"] = dataframe["W"] + dataframe["X"]
```

```python
dataframe
```

##### 3. Remoção de uma coluna

```python
dataframe.drop("new", axis=1) #axis=1 equivale a colunas | axis=0 equivale a linhas
```

##### 4. Cópia ou original?

A maioria, se não todas as funções do pandas não alteram efetivamente o dataset, elas apenas fazem uma cópia do mesmo, permitindo que você trabalhe com os dados modificados e os vizualize, mas o dataset original segue intacto, veja:

```python
dataframe
```

```python
dataframe.drop("new", axis=1)
```

```python
dataframe
```

##### 5. Alterações no dataframe original

```python
dataframe
```

```python
dataframe.drop("sum", axis=1, inplace=True)
```

```python
dataframe
```

### 6. Loc, Iloc e Filtros

### 1. Loc

```python
dataframe
```

```python
dataframe.loc["A"]
```

```python
dataframe.loc[["A"]]
```

```python
dataframe.loc[["A", "B"]]
```

```python
dataframe.loc[["A", "B"],"W"]
```

### 2. Iloc

```python
dataframe.iloc[3, 2]
```

```python
dataframe.iloc[0, 0]
```

```python
dataframe.iloc[:-1, :]
```

```python
dataframe.iloc[:-2, 1:]
```

```python
dataframe.iloc[1:4, 1:3]
```

```python
dataframe > 0
```

```python
dataframe[dataframe > 0]
```

```python
dataframe
```
