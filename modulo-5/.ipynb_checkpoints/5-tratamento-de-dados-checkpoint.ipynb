# Tratamento de dados ausentes

```python
import pandas as pd
import numpy as np
```

```python
dataframe = pd.DataFrame({ "A": [1, np.nan, 2],
                           "B": [5, np.nan, np.nan],
                           "C": [1, 2, 3]})
                           
```

```python
dataframe
```

```python
dataframe.dropna()
```

```python
dataframe.dropna(axis=1)
```

```python
dataframe.dropna(axis=1, thresh=2)
```

```python
dataframe.fillna("Valor")
```

```python
dataframe["A"].fillna(value=dataframe["A"].mean())
```

```python
dataframe
```

```python
dataframe.ffill()
```

```python
dataframe.bfill()
```
