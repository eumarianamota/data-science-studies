# Series Temporais

Series temporais são conjuntos de dados coletados de forma sequencial ao longo de tempo, geralmente em intervalos rgulares (dia, hora, minuto). Essas series são muito usadas na área de finanças para analisar padrões, tendências e prever o futuro.

```python
import pandas as pd
import datetime
```

```python
numero_de_dias = 100
```

```python
datas = pd.date_range(start="1/1/2024", periods=numero_de_dias)
```

```python
type(datas)
```

```python
dataframe = pd.DataFrame(range(numero_de_dias), columns=["number"], index=datas)
```

```python
dataframe
```

```python
dataframe.index
```

```python
dataframe.index[0]
```

```python
dataframe.index[2].day
```

```python
dataframe.index[3].month
```

```python
dataframe.index[3].year
```

```python
dataframe[dataframe.index.month == 4]
```

```python
dataframe[dataframe.index.day == 2]
```

```python
dataframe[dataframe.index > datetime.datetime(2024, 3, 30)]
```
