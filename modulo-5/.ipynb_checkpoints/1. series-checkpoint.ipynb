# Series

```python
import pandas as pd
import numpy as np
```

```python
labels = ['a', 'b', 'c']
minha_lista = [10, 20, 30]
arr = np.array([10, 20, 30])
d = {'a':10, 'b': 20, 'c':30}
```

```python
pd.Series(labels)
```

```python
pd.Series(data=labels, index=minha_lista)
```

```python
pd.Series(labels, minha_lista)
```

```python
pd.Series(d)
```
