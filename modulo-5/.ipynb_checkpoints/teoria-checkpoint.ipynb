## 1. Introdu√ß√£o 

1. Series: um nome diferente para coluna, √© um array que armazena todos os valores de uma coluna em uma tabela.
2. Dataframe: uma estrutura de dados que armazena dados em colunas(series) e linhas.

| :- | :- |
| dataframe.head() | Mostra as linhas do topo de uma tabela | 
| dataframe.tail()| Mostra as linhas do final de uma tabela |

| dataframef.memory_usage() | Mostra o uso de mem√≥ria de cada coluna de um dataframe | 
| Valores exclusivos | No pandas, valores exclusivos se referem √† elementos √∫nicos presentes em uma serie ou coluna |
| dataframe["col2"].unique() | Mostra um array com os elementos √∫nicos da coluna | 
| dataframe["col2"].nunique() | Mostra a quantidade de elementos √∫nicos da coluna | 
| dataframe["col2"].value_counts() | Mostra a quantidade de vezes que cada um dos elementos aparece na coluna |
| df["col1"].sum() | Mostra a soma dos valores da coluna selecionada |
| df["col1"].mean() | Mostra a m√©dia dos valores da coluna selecionada |
| df["col1"].product() | Mostra o resultado da multiplica√ß√£o de todos os valores da coluna (produto) |
| df["col1"].std() | Mostra o desvio padr√£o dos valores da coluna | 
| df["col1"].max() | Mostra o maior valor da coluna | 
| df["col1"].min() | Mostra o menor valor da coluna |
| df["col1"].idxmax() | Mostra o index do maior valor da coluna |
| df["col1"].idxmin() | Mostra o id do menor valor da coluna |

| dataframe["serie"] | Selecionando uma serie do dataframe |
| dataframe[["serie"]] | Selecionando uma serie do dataframe e deixando ela no formato dataframe | 
| dataframe[["serie1", "serie2"]] | Sele√ß√£o de m√∫ltiplas series no formato dataframe |

| iloc[] | O iloc √© uma ferramenta usada para encontar dados a partir do index das linhas e/ou colunas |

| dataframe.dropna(axis=1) | Remove todas as colunas que tem pelo menos um NAN |

| data.groupby("Classe").min() | Cria um agrupamento simples por classe e retorna o valor m√≠nimo de cada coluna |

### 2. Merge

```python
esquerda = pd.DataFrame({"key": ["k0", "k1", "k2", "k3"],
                         "A": ["A0", "A1", "A2", "A3"],
                         "B": ["B0", "B1", "B2", "B3"]})
direita = pd.DataFrame({"key": ["k0", "k1", "k2", "k3"],
                         "C": ["C0", "C1", "C2", "C3"],
                         "D": ["D0", "D1", "D2", "D3"]})
```

```python
esquerda
```

```python
direita
```

```python
pd.merge(esquerda, direita, on="key")
```

```python
esquerda = pd.DataFrame({"key1": ["k0", "k0", "k1", "k2"],
                         "key2": ["k0", "k1", "k0", "k1"],
                         "A": ["A0", "A1", "A2", "A3"],
                         "B": ["B0", "B1", "B2", "B3"]})
direita = pd.DataFrame({"key1": ["k0", "k1", "k1", "k2"],
                        "key2": ["k0", "k0", "k0", "k0"],
                         "C": ["C0", "C1", "C2", "C3"],
                         "D": ["D0", "D1", "D2", "D3"]})
```

```python
esquerda
```

```python
direita
```

```python
pd.merge(esquerda, direita, on=["key1", "key2"])
```

```python
pd.merge(esquerda, direita,how="left", on=["key1", "key2"])
```

```python
pd.merge(esquerda, direita, how="right", on=["key1", "key2"])
```

```python
pd.merge(esquerda, direita, how="outer", on=["key1", "key2"])
```

## 9. Series temporais
