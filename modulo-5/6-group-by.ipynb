# Group By

```python
import pandas as pd
```

```python
data = pd.DataFrame({ "Classe": ["Júnior", "Júnior", "Pleno", "Pleno", "Sênior", "Sênior"], 
                      "Nome": ["Jorge", "Carlos", "Roberta", "Mariana", "Alice", "Juliana"],
                      "Venda": [200, 120, 340, 124, 243, 350]})
```

```python
data
```

```python
data.groupby("Classe")["Venda"].sum()
```

```python
data.groupby("Classe").sum(numeric_only=True)
```

```python
data.groupby("Classe").mean(numeric_only=True)
```

```python
data.groupby("Classe").min()
```

```python
data.groupby("Classe").max()
```

```python
dataframe = data.copy()
```

```python
data
```

```python
dataframe
```

```python
dataframe["Venda"] = [150, 432, 190, 230, 410, 155]
```

```python
dataframe
```

```python
data
```

```python
dataframe2 = pd.concat([data, dataframe])
```

```python
dataframe2
```

```python
dataframe2.groupby(["Classe", "Nome"]).min()
```

```python
dataframe2.groupby(["Classe", "Nome"]).max()
```

```python
dataframe2.groupby(["Classe", "Nome"]).sum()
```

```python
dataframe2.groupby(["Classe", "Nome"]).mean()
```
