# Series

```python
import pandas as pd
import numpy as np
```

### 1. Series criação

#### 1. Criando diferentes estruturas de dados

```python
minha_lista_string = ['a', 'b', 'c']
minha_lista_number = [10, 20, 30]
meu_array = np.array([10, 20, 30])
meu_dicionario = {'a':10, 'b': 20, 'c':30}
```

#### 2. Criando uma serie pura

```python
serie = pd.Series([1, 2, 3, 4, 5])
```

```python
serie
```

#### 3. Criando uma serie com uma lista

```python
pd.Series(minha_lista_string)
```

#### 4. Criando uma serie definindo argumentos

```python
pd.Series(data=minha_lista_string, index=minha_lista_number)
```

```python
pd.Series(minha_lista_string, minha_lista_number)
```

#### 5. Criando uma serie com um dicionário

```python
pd.Series(meu_dicionario)
```

#### 6. Criando series e somando

```python
serie1 = pd.Series([1, 2, 3, 4], index=["Brasil", "Tailândia", "Islândia", "Nepal"])
```

```python
serie2 = pd.Series([1, 2, 3, 4], index=["Brasil", "Tailândia", "Islândia", "Groenlândia"])
```

```python
serie1
```

```python
serie1 + serie2 
```

#### 7. Acessando uma celula de serie pelo index

```python
serie1[["Brasil"]]
```

```python
serie1[["Brasil", "Islândia"]] 
```
