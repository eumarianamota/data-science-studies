### 3. √çndices multin√≠veis

```python
outside = ["Mulheres", "Mulheres", "Mulheres", "Homens", "Homens", "Homens"]
inside = ["Filial1", "Filial2", "Filial3", "Filial1", "Filial2", "Filial3", ]
hier_index = list(zip(outside, inside))
```

```python
hier_index
```

```python
hier_index = pd.MultiIndex.from_tuples(hier_index)
```

```python
dataframe2 = pd.DataFrame(np.random.randn(6, 2), index=hier_index, columns=["A", "B"])
```

```python
dataframe2
```

```python
dataframe2.loc["Mulheres"]
```

```python
dataframe2.loc["Mulheres"].loc["Filial1"]
```

```python
dataframe2.index
```

```python
dataframe2.index.names
```

```python
dataframe2.index.names = ["Grupo", "Filial"]
```

```python
dataframe2
```

```python
dataframe2.xs("Filial1", level="Filial")
```
