#### 1. utilizando o built-in method input(), crie um rpograma que receba a altura e o peso de uma pessoa e imprima na tela o IMC da mesma.

```python
height = float(input("Digite a sua altura: "))
weight = float(input("Digite o seu peso: "))
# type(height)
imc =  weight / height 
print("O seu IMC é {}.".format(f'{imc:.1f}'))
```

#### 4. Faça ump programa para uma loja de tintas. A pessoa informa a área em m² que deseja pintar, e o script calcula a quantidade de latas de tinta que a pessoa deve comprar e o valor. Considere que cada litro de tinta pinta 3m², que cada lata contém 18L de tinta e que custa R$58,00. 

```python
import math

area = float(input("Informe a área em m² que você deseja pintar."))
wall_paint_liters = area / 3
wall_paint_cans = math.ceil((wall_paint_liters / 18))
spent_money = wall_paint_cans * 58

print("Você deve comprar {} lata(s) de tinta, o que dá R$:{}".format(wall_paint_cans, f'{spent_money:.2f}'))
```

#### 5. Faça um programa que pergunte quantoo você ganha por hora e o número de horas trabalhadas por mês. Calcule e mostre o total do seu salário no referido mês, sabendo-se que são descontados 11% para o imposto de renda, 8% para o INSS e 5% para o sindicato. Faça um programa que nos dê:
    1. Salário bruto
    2. Quanto pagou ao INSS
    3. Quanto pagou ao sindicato
    4. Salário líquido

```python
earned_hours = float(input("Informe quanto você ganha por hora."))
worked_hours = float(input("Informe quantas horas você traalha por mês."))

gross_salary = earned_hours * worked_hours
net_salary = gross_salary

income_tax_percentage = 100 - 11
inss_tax_percentage = 100 - 8
union_tax_percentage = 100 - 5

net_salary = (income_tax_percentage * net_salary) / 100 #6230
income_tax = gross_salary - net_salary 

net_salary = (inss_tax_percentage * net_salary) / 100 #5731,60
inss_tax = (net_salary + income_tax) - net_salary

net_salary = (union_tax_percentage * net_salary) / 100
union_tax = (net_salary + income_tax + inss_tax) - net_salary

print("Salário bruto: {} \nINSS: {} \nSindicato: {} \nSalário líquido: {}".format(gross_salary, inss_tax, union_tax, net_salary))
```
