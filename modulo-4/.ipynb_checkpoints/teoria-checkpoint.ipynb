## 1.2. criação de atributos em uma classe

```python
class Dog:
    def __init__(self, breed): #Uma função de inicalização
        self.breed = breed
        self.age = 10
        print("O cachorro de {} anos é da raça {}.".format(self.age, self.breed))
```

```python
dog1 = Dog("Vira-lata")
```

## 1.3. Criação de métodos em uma classe

```python
class Dog:
    def __init__(self, breed): #Uma função de inicalização
        self.breed = breed
        self.age = 10
        print("O cachorro de {} anos é da raça {}.".format(self.age, self.breed))

    def getOld(self):
        self.age += 1
        return self.age
```

```python
dog1 = Dog("Husky")
dog1.getOld()
```

## 1.4. Parâmetros pré-definidos

```python
class Circle:
    def __init__(self, raio=1):
        self.raio = raio
        print("O raio do objeto é {}".format(self.raio))

    def calcula_area(self):
        return self.raio * self.raio * 3.14
```

```python
circle1 = Circle()
circle2 = Circle(2)
circle2.calcula_area()
```

## 1.5. Herança

```python
class Animal:
    def __init__(self):
        print("Animal criado!")
     
    def quem_sou_eu(self):
        print("Eu sou um animal!")

    def comer(self):
        print("Comendo...")

class Gato(Animal):
    def __init__(self):
        Animal.__init__(self)
        print("Gato criado!")
     
    def quem_sou_eu(self):
        print("Heyy, eu sou um gato!")
```

```python
animal = Animal()
```

```python
animal.comer()
```

```python
gato = Gato()
```

```python
gato.quem_sou_eu()
```

## 1.6. Métodos especiais

```python
class Book:
    def __init__(self, title, author, pages):
        print("The book was created")
        self.title = title
        self.author = author
        self.pages = pages 

    def __str__(self): #Caso você faça um print do objeto "print(book1)", isso vai ser retornado 
        return "Title: {} \nAuthor: {} \nPages: {}".format(self.title, self.author, self.pages)

    def __len__(self): #Define o tamanho que será retornado quando a função len for chamada "print(len(book1))", nesse caso, as páginas do livro
        return self.pages

    def __del__(self): #Define o que acontece antes da função "del(book1)" ser finalizada e deletar o objeto
        print("The book was destroyed")
```

```python
book = Book("A Amiga Genial", "Elena Ferrante", 302)
```

```python
print(book)
```

```python
print(len(book))
```

```python
del book
```
