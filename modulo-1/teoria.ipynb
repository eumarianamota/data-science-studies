## 1.9. Operadores de comparação

#### 1.9.1. Tabela de operadores de comparação

|  Operador | nome|  Descrição | Exemplo |
| :- | :- | :- | :- |
| == | igual a | avalia se os operandos são iguais | a == b |
| != | difernete de | avalia se os operandos são diferentes | a != b |
| > | maior que |avalia se o operando esquerdo é maior que o direito | 10 > 5 |
| < | menor que |avalia se o operando esquerde é menor que o direito | 5 < 10 |
| >= | maior ou igual a | avalia se o operando esquerdo é maior ou igual ao direito | 5 >= 5 |
| <= | menor ou igual a | avalia se o operando esquerdo é menor ou igual ao direito | 3 <= 5 |

```python
1 == 1
```

```python
1 != 2
```

```python
2 > 4
```

```python
2 < 4
```

```python
2 >= 2
```

```python
2 <= 2
```
