# Calculadora 

- 0: Soma
- 1: Subtração
- 2: Multiplicação
- 3: Divisão
- 4: Exponenciação

 Escolha uma operação que deseja realizar:

Qual o segundo valor?

(Resultado da operação)

Deseja fazer outra operação? 0 - SIM, 1 - NÂO
(Faz a limpeza da tela do terminal)
(Repete o processo se for sim e para se for não)

```python
from IPython.display import clear_output

def Clear():
    clear_output(wait=True)

def Sum(value1, value2):
    return value1 + value2


def Subtraction(value1, value2):
    return value1 - value2


def Multiplication(value1, value2):
    return value1 * value2


def Division(value1, value2):
    return value1 / value2


def Potentiation(value1, value2):
    return value1 ** value2


def Calculation(value1, value2, operation):
    if operation == 0:
        aftermath = Sum(value1, value2)
        return ("{} + {} = {}".format(value1, value2, aftermath))
    elif operation == 1:
        aftermath = Subtraction(value1, value2)
        return ("{} - {} = {}".format(value1, value2, aftermath))
    elif operation == 2:
        aftermath = Multiplication(value1, value2)
        return ("{} x {} = {}".format(value1, value2, aftermath))
    elif operation == 3:
        aftermath = Division(value1, value2)
        return ("{} % {} = {}".format(value1, value2, aftermath))
    elif operation == 4:
        aftermath = Potentiation(value1, value2)
        return ("{} ** {} = {}".format(value1, value2, aftermath))


def Calculator():
    print("0: Adição \n1: Subtração \n2: Multiplicação \n3: Divisão \n4: Potenciação")
    operation = int(input("Escolha a operação que deseja realizar: "))

    value1 = float(input("Qual o primeiro valor? "))
    value2 = float(input("Qual o segundo valor? "))

    result = Calculation(value1, value2, operation)
    print(result)

    continuation = int(input("Deseja fazer outra operação? 0: SIM 1:NÂO "))
    if continuation == 0:
        Clear()
        Calculator()
    else:
        print("Operação realizada com sucesso, até mais ver!")


Calculator()
```
