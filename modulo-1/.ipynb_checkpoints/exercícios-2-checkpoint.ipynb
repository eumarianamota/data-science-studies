# Exercícios - operadores de comparação e laço

#### 1. Faça um programa para a leitura de duas notas parciais de um aluno. O programa deve calcular a média alcançada por aluno e apresentar:
- A mensagem "Aprovado" se a média for maior ou igual a 7.
- A mensagem "Reprovado" se a média for menor que 7.
- A mensagem "Aprovado com distinção" se a média for igual a 10.

#### 2. Escreva um script que leia três números e mostre o maior e o menor deles

#### 3. Nome da vertical em escada. Modifique o programa anterior de forma a mostrar o nome em formato de escada.
M

MA

MAR

MARI

MARIA

MARIAN

MARIANA

#### 4. A série de Fibonacci é formada pela sequência 1, 1, 2, 3, 5, 8, 14, 21, 34, 55 (o próximo termo, a partir do terceiro, é sempre gerado a partir do somatório dos últimos dois). Faça um programa capaz de gerar uma série até o n-ésimo termo (onde o valor n deve ser inserido pelo usuário).

#### 5. Faça um programa que leia e valide as seguintes informações:
- Nome: maior que três caracteres 
- Idade: entre 0 e 150
- Salário: maior que zero
- Sexo: "f" ou "m"
- Estado civil: "s", "c", "v" ou "d"

#### 6. Faça um programa que peça um número inteiro e determine se ele é ou não um número primo. Um número primo é aquele que é divisível somente por ele mesmo e por 1. Dica: utilize o operador aritmético %, que retorna o resto da divisão de dois números.
