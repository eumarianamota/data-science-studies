## 1.1. Manipulação de números

| Expressão | Descrição | Resultado |
| :- | :- | :- |
| 2 + 3 | Adição de dois ou mais valores | 2 + 3 = 5 |
| 2 - 1 | Subtração de dois ou mais valores | 2 - 1 = 5 | 
| 2 * 2 | Multilicação de dois ou mais valores | 2 * 2 = 4 |
| 2 / 2 | Divisão de dois ou mais valores | 2 / = 1 |
| 2 ** 2 | Potenciação ou elevação de um número | 2 ** 3 = 4  |
| 9 ** (1/2) | Raiz quadrada de um número | 2 **(1/2) = 3 |

## 1.9. Operadores de comparação

#### 1.9.1. Tabela de operadores de comparação

|  Operador | nome|  Descrição | Exemplo |
| :- | :- | :- | :- |
| == | igual a | avalia se os operandos são iguais | a == b |
| != | difernete de | avalia se os operandos são diferentes | a != b |
| > | maior que |avalia se o operando esquerdo é maior que o direito | 10 > 5 |
| < | menor que |avalia se o operando esquerde é menor que o direito | 5 < 10 |
| >= | maior ou igual a | avalia se o operando esquerdo é maior ou igual ao direito | 5 >= 5 |
| <= | menor ou igual a | avalia se o operando esquerdo é menor ou igual ao direito | 3 <= 5 |

```python
1 == 1
```

```python
1 != 2
```

```python
2 > 4
```

```python
2 < 4
```

```python
2 >= 2
```

```python
2 <= 2
```

```python
 x = 0

while x <= 10:
    print("O valor de x é {}".format(x))
    x = x + 1
    print("Não passa pelo while")
else:
    print("Concluído")
```

```python
 x = 0

while x <= 10:
    print("O valor de x é {}".format(x))
    x = x + 1
    continue
    print("Não passa pelo while")
else:
    print("Concluído")
```

```python
 x = 0

while x <= 10:
    print("O valor de x é {}".format(x))
    x = x + 1
    break
else:
    print("Concluído")
```

## 1.15. Compreensão em listas 

```python
words = "string"
list_words = []

for w in words:
    list_words.append(w)
    
print(list_words)
```

```python
nome = "mariana da mota"
```

```python
lista = [x for x in nome]
print(lista)
```

```python
lista2 = [x**2 for x in range(1, 11)]
print(lista2)
```

```python
lista3 = [x for x in range(30) if x % 2 != 0]
print(lista3)
```

```python
lista3 = [str(x) for x in range(30) if x % 2 != 0]
print(lista3)
```

```python
y = 1

lista3 = [(x + y) for x in range(30) if x % 2 != 0]
print(lista3)
```

## 1.16. Funções

```python
def diga_ola():
    print("oie")

diga_ola()
```

```python
def diga_ola(name):
    print("oie {}".format(name))

diga_ola("mariana")
```

```python
def impar(numbers):
    impares = []
    for number in range(numbers + 1):
        if number % 2 != 0:
            impares.append(number)
    return impares


impar(15)
```
