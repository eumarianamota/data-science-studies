## 1. Try, except, finally

- O try except é usado quando você tem duvida quanto ao funcionamento de um determinado bloco de código, então coloca ele dentro de um try, se ele funcionar, beleza, se não funcionar, vai ser redirecionado para o except que é uma altenativa ao código principal. Dessa maneira, o erro que iria fazer o projeto parar não vai acontecer porque ele está protegido com o try.

```python
x = [1, 2, 3]
y = 1

try:
    x + y
except:
    print("deu erro")
```

```python
x = [1, 2, 3]
y = 1

try:
    x + y
except Exception as e:
    print("deu erro: {}".format(e))
```

- O finally permite que todo e qualquer código seja executado, independentemente da ocorrência anterior de um erro ou não.

```python
x = [1, 2, 3]
y = 1

try:
    x + y
except Exception as e:
    print("deu erro: {}".format(e))
finally:
    print("Olá")
```

```python
x = [1, 2, 3]
y = 1

try:
    y + y
except Exception as e:
    print("deu erro: {}".format(e))
finally:
    print("Olá")
```

## 2. Logging

### Introdução 

O logging é o processo de registro de eventos que ocorrem ao longo da execução de um software. Essa ferramenta é importe porque permite ao programador conhecer o funcionamento do seu código, localizar bugs e otimizar o seu script. 

| Tipo| Código | Descrição | 
| :- | :- | :- |
| Importação | import logging | Para fazer a importação da biblioteca |
| Criação | logging.basicConfig(filename="critical.log", level=logging.DEBUG, filemode="w", format=formatoX) | Para criar o arquivo de logs |
|Instanciação | log = logging.getLogger() | Para registrar os logs, primeiro, é necessário instanciar o objeto responsável por manipular esses registros com o método "logging.getLogger |
|Registro | log.info("Registro aqui") | faz o registro de um log do tipo info |
| Registro | log.error("Registro aqui") | Faz o registro de um log do tipo erro | 

### Niveis de logging




- Nem todas as mensagem de log são iguais. Ao definir um nível de registro em python usando o módulo padrão, você está informando à biblioteca que deseja lidar com todos os eventos desse nível em diante.
    - Exemplo: se você definir o nível do log para INFO, você irá incluir todas as mensagens de info, warning, error e critical, mas vai excluir as mensagens de debug e noset. Veja os níveis abaixo:
 
|  Nível | Valor numérico |
| :- | -: | 
| Critical | 50 |
| Error | 40 |
| Warning | 30 |
| Info | 20 |
| Debug | 10 |
| Noset | 0 |

### Ação

```python
import logging
```

```python
LOG_FORMAT = "%(levelname)s %(asctime)s - %(message)s"
```

```python
logging.basicConfig(filename="app.log", level=logging.DEBUG,filemode="w", format=LOG_FORMAT)
```

```python
log = logging.getLogger()
```

```python
log.info("Registrando a primeira informação desse log")
log.error("Registrando o primeiro erro desse arquivo de logs")
```
