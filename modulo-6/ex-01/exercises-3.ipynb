# Transformação e agrupamentos simples

```python
import pandas as pd
```

```python
df = pd.read_csv("2004-2021.tsv", sep="\t")
```

```python
df.head()
```

### 1. Conversão de Datas: Converta a coluna "DATA FINAL" para o formato datetime do Pandas.

```python
df["DATA FINAL"] = pd.to_datetime(df["DATA FINAL"])
```

```python
type(df["DATA FINAL"][0])
```

### 2. Extração de Ano: Crie uma nova coluna chamada "ANO" extraída apenas da "DATA FINAL".

```python
df["ANO"] = df["DATA FINAL"].apply(lambda x: x.year)
```

```python
df[["DATA FINAL", "ANO"]].head()
```

### 3. Renomeação: Altere o nome das colunas "REGIÃO" para "REGION" e "ESTADO" para "STATE" (apenas para praticar a função de renomear).

```python
df.rename(columns={"REGIÃO": "REGION", "ESTADO": "STATE"}).head(1)
```

### 4. Média por Região: Qual é o preço médio de revenda da gasolina para cada uma das 5 regiões do Brasil? (Dica: use groupby).

```python
df.groupby("REGIÃO")["PREÇO MÉDIO REVENDA"].mean()
```

### 5. Volume de Dados: Conte quantos registros existem para cada estado na base de dados.

```python
df.groupby("REGIÃO")["ESTADO"].value_counts()
```

### 6. Unidades de Medida: Identifique quais produtos são vendidos em "R/l" e quais são vendidos em "R$13kg".

```python
df[df["UNIDADE DE MEDIDA"].isin(["R$/13Kg"])]["PRODUTO"].value_counts()
```

```python
df[df["UNIDADE DE MEDIDA"].isin(["R$/l"])]["PRODUTO"].value_counts()
```
