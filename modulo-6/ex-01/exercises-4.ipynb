# Filtros, Seleção e Limpeza

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("2004-2021.tsv", sep="\t")
```

### 1. Visão Final: Exiba as últimas 7 linhas do conjunto de dados para verificar o final da série histórica.

```python
df.tail(7)
```

### 2. Foco no Norte: Crie um filtro que retorne apenas os dados do estado do Amazonas.

```python
df[df["ESTADO"] == "AMAZONAS"]
```

### 3. Análise de Gás: Qual é a média de postos pesquisados especificamente para o produto "GNV"?

```python
df[df["PRODUTO"] == "GNV"]["NÚMERO DE POSTOS PESQUISADOS"].mean()
```

### 4. Seleção de Colunas: Crie um novo DataFrame contendo apenas as colunas: DATA FINAL, ESTADO, PRODUTO e PREÇO MÉDIO REVENDA.

```python
df_new = df[["DATA FINAL", "PRODUTO", "PREÇO MÉDIO REVENDA"]].copy()
```

```python
df_new.head()
```

### 5. Máximo do Diesel: Identifique qual foi o maior "PREÇO MÁXIMO REVENDA" registrado para o "ÓLEO DIESEL S10" em todo o Brasil.

```python
df[df["PRODUTO"] == "ÓLEO DIESEL S10"]["PREÇO MÁXIMO REVENDA"].max()
```

### 6. Preços Baixos: Filtre todos os registros onde o "PREÇO MÉDIO REVENDA" foi menor que R$ 2,00. 

```python
df[df["PREÇO MÉDIO REVENDA"] < 2]
```

### 7. Frequência por Estado: Conte quantas vezes cada estado aparece no conjunto de dados (frequência de registros).

```python
df["ESTADO"].value_counts()
```

### 8. Coluna de Mês: Crie uma nova coluna chamada "MES_NUM" extraída da coluna DATA FINAL (apenas o número do mês).

```python
df["DATA FINAL"] = pd.to_datetime(df["DATA FINAL"])
```

```python
type(df["DATA FINAL"][0])
```

```python
df["MES"] = df["DATA FINAL"].apply(lambda x: x.month)
```

```python
type(df["MES"][0])
```

```python
df["MES"].value_counts()
```

### 9. Recorte Regional: Filtre o DataFrame para exibir apenas dados pertencentes à região "SUDESTE".

```python
df[df["REGIÃO"] == "SUDESTE"]
```

### 10. Instabilidade de Preço: Localize os 5 registros que possuem o maior "DESVIO PADRÃO REVENDA".

```python
df["DESVIO PADRÃO REVENDA"].nlargest(5)
```

### 11. Tipos de Medida: Liste todos os valores únicos da coluna UNIDADE DE MEDIDA.

```python
df["UNIDADE DE MEDIDA"].value_counts()
```

### 12. Ano Específico: Filtre os dados do produto "GASOLINA COMUM" referentes apenas ao ano de 2015.

```python
df["YEAR"] = df["DATA FINAL"].apply(lambda x: x.year)
```

```python
df[(df["PRODUTO"] == "GASOLINA COMUM") & (df["YEAR"] == 2015)]
```

### 13. Distribuição na Bahia: Calcule o "PREÇO MÉDIO DISTRIBUIÇÃO" apenas para o estado da Bahia.

```python
df["PREÇO MÉDIO DISTRIBUIÇÃO"] = pd.to_numeric(df["PREÇO MÉDIO DISTRIBUIÇÃO"], errors="coerce").replace(-99999.0, np.nan)
```

```python
df[df["ESTADO"] == "BAHIA"]["PREÇO MÉDIO DISTRIBUIÇÃO"].mean()
```

### 14. Barreira dos 7 Reais: Identifique em quais datas e estados o "PREÇO MÁXIMO REVENDA" atingiu ou ultrapassou R$ 7,00.

```python
df[df["PREÇO MÁXIMO REVENDA"] > 7][["DATA FINAL", "ESTADO", "PREÇO MÁXIMO REVENDA"]]
```

### 15. Ordenação por Amostra: Ordene o DataFrame pelo NÚMERO DE POSTOS PESQUISADOS de forma crescente.

```python
df.sort_values(by="NÚMERO DE POSTOS PESQUISADOS")
```

### 16. Filtro Sulista: Crie um filtro para encontrar registros de "GLP" (Gás de Cozinha) apenas na região "SUL". 

```python
df[(df["PRODUTO"] == "GLP" ) & (df["REGIÃO"] == "SUL")]
```

### 17. Valores Nulos: Verifique se existem valores nulos (NaN) na coluna PREÇO MÉDIO DISTRIBUIÇÃO.

```python
df["PREÇO MÉDIO DISTRIBUIÇÃO"].isnull().sum()
```

### 18. Amplitude de Preço: Crie uma nova coluna chamada "AMPLITUDE_REVENDA" que seja a diferença entre o preço máximo e o preço mínimo de revenda.

```python
df["AMPLITUDE REVENDA"] = df["PREÇO MÁXIMO REVENDA"] - df["PREÇO MÍNIMO REVENDA"]
```
